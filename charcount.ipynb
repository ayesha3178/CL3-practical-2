```python
# Character Count Program

file = open("input.txt", "r")

data = file.read()

count = {}

for ch in data:

    if ch != " " and ch != "\n":

        if ch in count:
            count[ch] += 1
        else:
            count[ch] = 1

print("Character Count:\n")

for key, value in count.items():
    print(key, ":", value)
```
